```python
cols=['Disease','Symptom_1','Symptom_2','Symptom_3']
```

```python
def get_weights(symptom):
    ind=int(d2[d2['Symptom']==symptom].index.values)
    return d2.iloc[ind]['weight']
```

```python
def remove_space(string):
    return string.replace(" ", "")
```

```python
w1=[]
w2=[]
w3=[]
s=[]
for i in range(4920):
    s1=remove_space(df.iloc[i]['Symptom_1'])
    s2=remove_space(df.iloc[i]['Symptom_2'])
    s3=remove_space(df.iloc[i]['Symptom_3'])
    w1.append(get_weights(s1))
    w2.append(get_weights(s2))
    w3.append(get_weights(s3))    
```

```python
df['sw_1']=w1
df['sw_2']=w2
df['sw_3']=w3
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 1000)
```

```python
rf.fit(x_train,y_train)
```

```python
predictions = rf.predict(x_test)
actual=y_test
predictions=list(predictions.reshape(-1,1))
predicted=[]
for i in predictions:
    for j in i:
        predicted.append(round(j))
actual=list(actual)
count=0
for i in range(len(actual)):
    if(actual[i]==predicted[i]):
        count+=1
accuracy_rf=count/len(actual)
print(str(round((accuracy_rf*100),2))+' %')
```

```python
def random_forest_predict(l):
    for i in (rf.predict([l])):
        return int(i)
```

```python
def rfp(pre):
    final=[]
    wts=[]
    for i in pre:
        wts.append(get_weights(i))
    # small fix needed
    pre[0]=dic1[pre[0]]
    pre[1]=dic2[pre[1]]
    pre[2]=dic3[pre[2]]
    final=pre+wts
    sc=StandardScaler().fit(df[cols_to_normalize])
    final=list((sc.transform([final]))[0])
    #numerical=int(random_forest_predict(final))
    #f=le_y.inverse_transform([numerical])[0]
    numerical=random_forest_predict(final)
    return (le_y.inverse_transform([numerical])[0])
```

```python
def knnp(pre):
    final=[]
    wts=[]
    for i in pre:
        wts.append(get_weights(i))
    # small fix needed
    pre[0]=dic1[pre[0]]
    pre[1]=dic2[pre[1]]
    pre[2]=dic3[pre[2]]
    final=pre+wts
    sc=StandardScaler().fit(df[cols_to_normalize])
    final=list((sc.transform([final]))[0])
    numerical=knn_predict(final)
    return (le_y.inverse_transform([numerical])[0])
    
```
